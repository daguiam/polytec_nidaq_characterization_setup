```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Define system parameters
m = 1.0     # mass (kg)
k = 10.0    # spring constant (N/m)
c = 1.0     # damping coefficient (Ns/m)

# Define initial conditions
x0 = 1.0    # initial displacement (m)
v0 = 0.0    # initial velocity (m/s)

# Define simulation parameters
dt = 0.01   # time step (s)
tmax = 10.0 # maximum time (s)

# Define lumped model equations
def lumped_model(state, t, u=0):
    x, v = state
    dxdt = v
    dvdt = -(k/m)*x - (c/m)*v + (1/m)*u
    return [dxdt, dvdt]

# Initialize simulation variables
t = np.arange(0, tmax+dt, dt)
x = np.zeros_like(t)
v = np.zeros_like(t)
x[0] = x0
v[0] = v0


# Define input signal
f = 1.0     # input frequency (Hz)
u = np.sin(2*np.pi*f*t)


# Run simulation using the lumped model
for i in range(1, len(t)):
    state = [x[i-1], v[i-1]]
    dxdt, dvdt = lumped_model(state, t[i], u[i])
    x[i] = x[i-1] + dxdt*dt
    v[i] = v[i-1] + dvdt*dt

# Compute frequency response
w, mag, phase = signal.bode(([1/m], [1/m, c/m, k/m]), w=2*np.pi*f)

# Plot simulation results and frequency response
fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(t, x, label='Displacement')
ax[0].plot(t, v, label='Velocity')
ax[0].set_xlabel('Time (s)')
ax[0].set_ylabel('Response')
ax[0].legend()

ax[1].semilogx(w/(2*np.pi), mag, label='Magnitude')
ax[1].semilogx(w/(2*np.pi), phase, label='Phase')
ax[1].set_xlabel('Frequency (Hz)')
ax[1].set_ylabel('Frequency Response (dB)')
ax[1].legend()

plt.show()
```
